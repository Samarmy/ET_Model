```python
csvs = glob.glob('zips/*.csv')
```

```python
for i in stations:
    Evapo_Dict = {}
    for j in date_list:
        day = str(j.day)
        month = str(j.month)
        year = str(j.year)
        if (len(day) == 1):
            day = '0' + day
        if (len(month) == 1):
            month = '0' + month
        Evapo_Dict[month + '_' + day + "_" + year] = []
    for j in csvs:
        if i.lower() in j.lower():
            with open(j, "r") as csvfile:
                csvreader = csv.DictReader(csvfile)
                for ind, row in enumerate(csvreader):
                    if (ind == 1):
                        if ("LE" in list(iter(row.values()))[1]):
                            LE_Index = list(iter(row.values()))[1].index("LE")
                        elif ("LE_1_1_1" in list(iter(row.values()))[1]):
                            LE_Index = list(iter(row.values()))[1].index("LE_1_1_1")
                        else:
                            print("ERROR CSV")
                            print(j)
                            break
                    if (ind<2):
                        continue
                    for ind2, row2 in enumerate(row.values()):
                        if (ind2 == 0):
                            year = int(row2[:4])
                            date_key = row2[4:6] + '_' + row2[6:8] + '_' + row2[:4]
                        else:
                            if ((year != 2015) and (year != 2016) and (year != 2017)):
                                continue
                            le = float(row2[LE_Index])
                            if (le != -9999):
                                Evapo_Dict[date_key].append(le)
    for key, value in Evapo_Dict.items():
        if (len(value) > 0):
            et_values = [(x/28.356) for x in value]
            Evapo_Dict[key] = (sum(et_values)/len(et_values))
        else:
            Evapo_Dict[key] = -9999
            
    with open('ET/' + str(i) + '.csv', 'w', newline='') as csvfile:
        fieldnames = ['DATE', 'ET']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for key, value in Evapo_Dict.items():
            writer.writerow({'DATE': key, 'ET': value})
                    
```
